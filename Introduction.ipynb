# AI vs ML vs DL

AI --> ML(subset of AI) --> DL(subset of ML)

AI - Artificial Intelligence is a branch of Computer Science that is concerned with building smart and intelligent machines 

Non - Intelligent Machines - Watch , Bike
Intelligent Machines - Autonomous Car , google assistant 

ML - Machine Learning is a technique to implement AI that can learn from data by themselves without being explicitly programmed.

DL - Subfield of ML that uses artificial neural networks to learn from data.









## Types Of Machine Learning 

1. Supervised Learning - In this , the machine learning algorithm learns from the labelled data.
 

2. Unsupervised Learning- In this ,  the machine learning algorithm learns from unlabelled data.


3. Reinforcement Learning - An area of ML concerned with how intelligent agents take actions in an environment to maximize its rewards.(Environment , Agent, Action, Reward)

### Supervised Learning and its Types 

1. Classification - It is about predicting a class or discrete values Eg: Male or Female , True or False .
Algorithms - 
Decision Tree
Random Forest 
K-nearest Neighbour


2. Regression - It is about predicting a quantity or continuous values Eg : Salary : age ; price. 
Algorithms- 
Logistic Regression 
Polynomial Regression
Support Vector Machines 





### Unsupervised Learning and its types 

The ML algo learns from unlabelled data.

Types :

1. Clustering - An unsupervised learning task which involves grouping the similar data points.

2. Association - An unsupervised task that is used tom find important relationship between data points.

Algorithms - 

- K-means Clustering 
- Hierarchical Clustering 
- Principal Component Analysis (PCA)
- Apriori
- Eclat 



### Deep Learning
