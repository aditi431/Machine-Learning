## Matplotlib : 

- Useful for making Plots 

There is alot of raw data drawing meaning insights from that data is tough so through ploting those data we can draw insights.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100) # evenly spaced 100 values that lies b/w 0 to 100
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# Plotting the data

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# Adding title x-axis y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# Adding title x-axis z-axis labels
plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('cos value')
plt.title('cosine wave')
plt.show()
```

```python
# Parabola

x = np.linspace(-10,10,20)
y = x**2 # equation of parabola
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g')
plt.plot(x,np.cos(x),'r--')
```

```python
# Bar plot 

# For data visualization


fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
# Piechart 

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
# Scatter plot 
x = np.linspace(0,10,10)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()



```

```python
# 3D scatter plot

fig = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
