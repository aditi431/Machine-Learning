# NumPy - Numerical Python 

Advantages of Numpy Arrays over list and tuples: 

    1. Allows several Mathematical Operations
    2. Faster Operations

```python
import numpy as np
```

```python
from time import process_time
```

```python
# Time taken by a list
python_list = [i for i in range(10000)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end_time  = process_time()

print(end_time - start_time)
```

```python
# Option 1: Wrap generator in a list
np_array = np.array([i for i in range(10000)])

# Option 2: Use NumPy's arange (best)
np_array = np.arange(10000)

start_time = process_time()
np_array += 5
end_time = process_time()

print(end_time - start_time)
```

```python
np_arr = np.array([1,2,3,4,5])
print(np_arr)
type(np_arr)
```

```python
# 1 dim array

a = np.array([1,2,3,4,5])
print(a)
print(a.shape) 
print(a.ndim)
```

```python
# 2 dim arrays

b = np.array([(1,2,3,4),(5,6,7,8)])
print(b)
print(b.shape)
print(b.ndim)
```

```python
c = np.array([(1,2,3,4),(5,6,7,8)],dtype=float)
print(c)
```

```python
# Initial Placeholders in numpy arrays

# numpy array of zeroes 

x = np.zeros((4,5))
print(x)
```

```python
# numpy array of ones

x = np.ones((3,3))
print(x)
```

```python

# array of a particular value

x = np.full((5,4),5)
print(x)
```

```python
# identity matrix 

x = np.eye(4)
print(x)
```

```python

# numpy array with random values

x = np.random.random((3,4)) # all values is between 0 and 1
print(x)

```

```python

# random integer value arrays within a specific range

x = np.random.randint(10,100,(3,5)) # range - start,end dim-(3,5)(row,col)
print(x)
```

```python
# array of evenly spaced value --> Specifying number of values

x = np.linspace(10,30,6) # values between 10 - 30 , 6 values
print(x)
```

```python
# array of evenly spaced value --> specifying the step

x = np.arange(10,30,5) # range (10,30) step (5)
print(x)
```

```python
# convert list to array

list1 = [10,20,30,40,50]
np_array = np.asarray(list1)
print(np_array)
type(np_array)
```

```python
# Analysing numpy array

c = np.random.randint(10,90,(5,5)) # value from 10-90 , 5x5 matrix
print(c)

print(c.shape) # rows columns
print(c.ndim)  # dimensions
print(c.size)  # no. of elements
print(c.dtype) # datatype of values present in arr
```

```python
# Mathematical Operations on numpy arrays 

# Addition of list - concatenate elements of lists or joins two lists 

# But in numpy arrays elements get actually added.

a = np.arange(10,30,5)
print(a)
b = np.arange(30,50,5)
print(b)

print(a+b)

```

```python
a = np.random.randint(10,100,(3,4))
print(a)
b = np.random.randint(10,100,(3,4))
print(b)
print(a+b)
```

```python
print(a-b)
print(a*b)
print(a/b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
# Array manipulation

array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)

trans =np.transpose(array)
print(trans)
print(trans.shape)
```

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)

trans2 = array.T
print(trans2)
print(trans2.shape)
```
